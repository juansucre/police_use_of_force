# Injury Level vs Resistance vs Race

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
civilian_df = pd.read_pickle('civilians_df.pkl')
```

```python
officer_df = pd.read_pickle('officer_df.pkl')
```

```python
injured_civs_df = civilian_df[civilian_df['Injured'] == True]
```

```python
injured_civs_df.head()
```

```python
death_df = injured_civs_df[injured_civs_df['Injury_Level'] == 'Death']
injury_df = injured_civs_df[injured_civs_df['Injury_Level'] == 'Injury']
serious_injury_df = injured_civs_df[injured_civs_df['Injury_Level'] == 'Serious bodily injury']
```

## Injury Level: Death -  Was there resistance? What type?

```python
black_deaths = death_df[death_df['Race_Ethnic_Group'] == 'Black']
white_deaths = death_df[death_df['Race_Ethnic_Group'] == 'White']
hispanic_deaths = death_df[death_df['Race_Ethnic_Group'] == 'Hispanic']
asian_deaths = death_df[death_df['Race_Ethnic_Group'] == 'Asian / Pacific Islander']

wbha_death_df = pd.concat([black_deaths, white_deaths, hispanic_deaths, asian_deaths])

print('Total black deaths:', len(black_deaths))
print('Total white deaths:', len(white_deaths))
print('Total hispanic deaths:', len(hispanic_deaths))
print('Total asian deaths:', len(asian_deaths))
```

```python
armed_resisted_injured_cols = ['CIVILIAN_Resisted', 'CIVILIAN_Resistance_Type', 
                               'CIVILIAN_Perceived_Armed', 'CIVILIAN_Confirmed_Armed']
```

### Black, Hispanic, White and Asian

```python
fig = plt.figure(figsize=(15,70))
for i, col in enumerate(armed_resisted_injured_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col, hue='Race_Ethnic_Group', data=wbha_death_df)

fig.tight_layout()
```

### Black

```python
fig = plt.figure(figsize=(15,70))
for i, col in enumerate(armed_resisted_injured_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col, data=black_deaths)

fig.tight_layout()
```

### White

```python
fig = plt.figure(figsize=(15,70))
for i, col in enumerate(armed_resisted_injured_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col, data=white_deaths)

fig.tight_layout()
```

### Hispanic

```python
fig = plt.figure(figsize=(15,70))
for i, col in enumerate(armed_resisted_injured_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col, data=hispanic_deaths)

fig.tight_layout()
```

### Asian / Pacific Islander

```python
fig = plt.figure(figsize=(15,70))
for i, col in enumerate(armed_resisted_injured_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col, data=asian_deaths)

fig.tight_layout()
```
