## Initial EDA Civilian-Officer Data
---
The Civilian-Officer dataset consists of details on the individuals involved in an incident and the outcome of said incident.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

#### Load data into DataFrames

```python
civ_off = pd.read_csv('./URSUS/URSUS_Civilian-Officer_2016.csv')
```

```python
civ_off_df = pd.DataFrame(civ_off)
```

#### Inspect shape and dtypes

```python
civ_off_df.shape
```

```python
civ_off_df.head()
```

The `Incident_ID` and `ORI` (The number for the reporting Agency) columns are irrelevant, so we can drop them.

```python
civ_off_df.head()
```

### Check for and fill nulls/nans

```python
civ_off_df.info()
```

As can be seen above, there are many columns with `Nan` values. Given that they are all categorical, I will assume that the `Nan`'s mean "not available" or simply that data was not collected for that category for that incident, and therefore will fill the nans with `na`'s.

```python
for col in civ_off_df.columns:
    mask = civ_off_df[col].isnull()
    civ_off_df[col] = civ_off_df[col].mask(mask, 'na')
```

```python
civ_off_df.isnull().sum().sum()
```

### Category types

```python
civ_off_df.dtypes
```

```python
demographic_cols = ['Gender',
                    'Race_Ethnic_Group',
                    'Asian_Race',
                    'Age']
```

```python
injury_cols = ['Injured',
               'Injury_Level',
               'Medical_Aid',
               'Injury_From_Preexisting_Condition',
               'Order_Of_Force_Specified',
               'Received_Force']
```

```python
civ_incident_cols = ['CIVILIAN_Assaulted_Officer',
                     'CIVILIAN_Perceived_Armed',
                     'CIVILIAN_Perceived_Armed_Weapon',
                     'CIVILIAN_Firearm_Type',
                     'CIVILIAN_Resisted',
                     'CIVILIAN_Resistance_Type',
                     'CIVILIAN_Custody_Status',
                     'CIVILIAN_Confirmed_Armed',
                     'CIVILIAN_Confirmed_Armed_Weapon',
                     'CIVILIAN_K12_Type',
                     'DISCHARGE_OF_FIREARM_INDIVIDUAL',
                     'DISCHARGE_OF_FIREARM_INCIDENT']
```

```python
off_incident_cols = ['OFFICER_Officer_Used_Force',
                     'OFFICER_Officer_Used_Force_Reason',
                     'OFFICER_On_Duty',
                     'OFFICER_Dress',
                     'DISCHARGE_OF_FIREARM_INDIVIDUAL',
                     'DISCHARGE_OF_FIREARM_INCIDENT']
```

```python
civ_off_df = civ_off_df.drop(['ORI', 'Asian_Race'], axis=1)
```

```python
civ_off_df.to_pickle('./dataframes/civ_off_df.pkl')
```

### Visual EDA on Civilian Data

```python
civilians_mask = civ_off_df['Civilian_Officer'] == 'Civilian'
```

```python
civilians_df = civ_off_df[civilians_mask]
```

#### Demographic

```python
fig = plt.figure(figsize=(15,50))
for i, col in enumerate(demographic_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col,data=civilians_df, orient='h')

fig.tight_layout()
```

Something appeared to have gone wrong when reading data and the number `10` was confused for the month `October`. To fix this, I set all the `17-Oct`'s as the correct `10-17`.

```python
oct_mask = civ_off_df['Age'] == '17-Oct'
```

```python
civ_off_df['Age'] = civ_off_df['Age'].mask(oct_mask, '10-17')
```

```python
oct_mask_civ = civilians_df['Age'] == '17-Oct'
```

```python
civilians_df['Age'] = civilians_df['Age'].mask(oct_mask_civ, '10-17')
```

```python
oct_mask_civ.sum()
```

```python
civilians_df.to_pickle('./civilians_df.pkl')
```

#### Incident details

```python
fig = plt.figure(figsize=(15,50))
for i, col in enumerate(civ_incident_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col,data=civilians_df, orient='h')

fig.tight_layout()
```

#### Injury/Medical/Force

```python
fig = plt.figure(figsize=(15,50))
for i, col in enumerate(injury_cols):
    fig.add_subplot(10,2,i+1)
    sns.countplot(y=col,data=civilians_df, orient='h')

fig.tight_layout()
```

# Civilian Race vs. Perceived Weapon

```python
white_civs = civilians_df[civilians_df['Race_Ethnic_Group'] == 'White']
black_civs = civilians_df[civilians_df['Race_Ethnic_Group'] == 'Black']
hispanic_civs = civilians_df[civilians_df['Race_Ethnic_Group'] == 'Hispanic']
asianpac_civs = civilians_df[civilians_df['Race_Ethnic_Group'] == 'Asian / Pacific Islander']

wbh_df = pd.concat([white_civs, black_civs, hispanic_civs, asianpac_civs])
```
