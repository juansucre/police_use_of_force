```python
fig, ax = plt.subplots(figsize=(10,8))

sns.barplot(ax=ax, y='%', x=civ_race_df.index[:4], data=civ_race_df[:4], palette="Blues_d")
plt.xlabel('Racial/Ethnic Group')
plt.ylabel('Percentage (%)')
```

```python
counties_pop = pd.read_csv('./dataframes/Untitled spreadsheet - Sheet1.csv')
```

```python
counties_pop.sort_values('%', ascending=False)
```

```python
incidents_county = pd.DataFrame(incident_df['County'].value_counts())

incidents_county.columns = ['# of incidents']

incidents_county['%'] = [(x/incidents_county['# of incidents'].sum())*100 for x in incidents_county['# of incidents']]

incidents_county['County'] = incidents_county.index

incidents_county['County'] = incidents_county['County'].str.replace('County', '')

incidents_county.columns = ['# of incidents', '% of incidents', 'County']

incidents_county = incidents_county.sort_values(by='County')
```

```python
counties_pop['Population'].values
```

```python
incidents_county['Population'] = counties_pop['Population'].values
```

```python
incidents_county['% of Pop'] = counties_pop['%'].values
```

```python
incidents_county = incidents_county.drop(['County'], axis=1)
```

```python
incidents_county.head()
```

```python
difference = []

for idx, row in incidents_county.iterrows():
    difference.append(row['% of incidents'] - row['% of Pop'])
```

```python
incidents_county['% difference'] = difference
```
