# All Incidents of Use of Force

```python
all_incidents_map.save('./dataframes/all_incidents_map.html')
```

```python
all_incidents_map = folium.Map((36.7783, -119.4179), zoom_start=6)
```

```python
coord_list_all = []
```

```python
for index, row in civilian_df.iterrows():
    coord_list_all.append([row['Loc_Latitude'], row['Loc_Longitude']])
```
