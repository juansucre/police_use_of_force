```python
civilian_df = pd.read_pickle('./dataframes/civilians_df.pkl')
```

```python
officer_df = pd.read_pickle('./dataframes/officer_df.pkl')
```

```python
civ_cols = ['CIVILIAN_Mental_Status', 'CIVILIAN_Assaulted_Officer',
            'CIVILIAN_Highest_Charge', 'CIVILIAN_Crime_Qualifier',
            'CIVILIAN_Perceived_Armed', 'CIVILIAN_Perceived_Armed_Weapon',
            'CIVILIAN_Firearm_Type', 'CIVILIAN_Resisted',
            'CIVILIAN_Resistance_Type', 'CIVILIAN_Custody_Status',
            'CIVILIAN_Confirmed_Armed', 'CIVILIAN_Confirmed_Armed_Weapon',
            'CIVILIAN_K12_Type']

off_cols = ['OFFICER_Officer_Used_Force',
            'OFFICER_Officer_Used_Force_Reason', 'OFFICER_On_Duty',
            'OFFICER_Dress']

civilian_df = civilian_df.drop(off_cols, axis=1)

officer_df = officer_df.drop(civ_cols, axis=1)
```

```python
civilian_df.to_pickle('./dataframes/civilian_df.pkl')
```

```python
officer_df.to_pickle('./dataframes/officer_df.pkl')
```

```python
serious_injured_civilian_df = civilian_df[civilian_df['Injury_Level'] == 'Serious bodily injury']
```

```python
sns.countserious_injured_civilian_df[]
```

```python
serious_injured_civ_ids = list(serious_injured_civilian_df['Incident_ID'].unique())
```

```python
serious_injured_civ_officers_involved_df = pd.DataFrame()

for id_ in serious_injured_civ_ids:
    df = officer_df[officer_df['Incident_ID'] == id_]
    serious_injured_civ_officers_involved_df = serious_injured_civ_officers_involved_df.append(df)
```

```python
list(serious_injured_civ_officers_involved_df['Age'].unique())
```

```python
ser_inj_off_inv_age_count = {}

for x in list(serious_injured_civ_officers_involved_df['Age'].unique()):
    ser_inj_off_inv_age_count[x] = (serious_injured_civ_officers_involved_df['Age'] == x).value_counts()[1]
    
```

```python
ser_inj_off_inv_age_count['61-65'] = 0
```

```python
ser_inj_off_inv_age_count
```

```python
all_off_age_count = {}

for x in list(officer_df['Age'].unique()):
    all_off_age_count[x] = (officer_df['Age'] == x).value_counts()[1]
    
```

```python
all_off_age_count['51-55']
```

```python
percentage_dict = {}
for key in all_off_age_count:
    percentage_dict[key] = ser_inj_off_inv_age_count[key] / all_off_age_count[key]
```

```python
percentage_dict
```
